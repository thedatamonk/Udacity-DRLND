### 2. Examine the State and Action Spaces

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents in the environment
print('Number of agents:', len(env_info.agents))

# number of actions
action_size = brain.vector_action_space_size
print('Number of actions:', action_size)

# examine the state space 
state = env_info.vector_observations[0]
print('States look like:', state)
state_size = len(state)
print('States have length:', state_size)
````

````python
# printing important variables of the environment (slightly different from the OpenAI environment)
print ("End of environment or not", env_info.local_done[0])
print ("Current state (37 dimension state) is ", env_info.vector_observations[0])
print ("Current state (visual represention) is ", env_info.visual_observations)
print ("Current action is ", env_info.previous_text_actions[0])
print ("Reward is ", env_info.rewards[0])
````

### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Note that **in this coding environment, you will not be able to watch the agent while it is training**, and you should set `train_mode=True` to restart the environment.

````python
env_info = env.reset(train_mode=True)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

### VanillaDQN with Experience replay and Fixed Q-targets

````python
from timeit import default_timer as timer
from datetime import timedelta
from collections import deque
import torch
````

````python
from VanillaDQN import Agent

agent = Agent(state_size=37, action_size=4, seed=0)
````

````python
def dqn(n_episodes = 2000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995):
        start = timer()    # timer starts for training
        scores = []        # list to contain the scores from each episode
        scores_window = deque(maxlen=100)   # last 100 scores
        scores_mean = []                    # list that scores mean of last 100 episodes
        eps = eps_start
        
        for i_episode in range(1, n_episodes+1):
            state = env.reset(train_mode=True)[brain_name].vector_observations[0]
            score = 0
            for t in range(max_t):
                
                action = agent.act(state, eps)
                env_info= env.step(action)[brain_name]
                next_state = env_info.vector_observations[0]
                reward = env_info.rewards[0]
                done = env_info.local_done[0]
                
                agent.step(state, action, reward, next_state, done)
                
                state = next_state
                score += reward
                if done:
                    break
            scores_window.append(score)    # save most recent score
            scores.append(score)           # save most recent score
            eps = max(eps_end, eps_decay*eps)  # decrease epsilon
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
            
            if i_episode % 100 == 0:
                print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            if i_episode >= 100:
                scores_mean.append(np.mean(scores_window))
            if np.mean(scores_window) >= 13.0:     # environment solved
                print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
                torch.save(agent.qnetwork_local.state_dict(), 'checkpoint_dqn.pth')
                break

        return scores, scores_mean, agent
````

````python
scores, scores_mean, agent = dqn()        
````

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# plotting per episode score achieved by the agent
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.title('DQN agent')
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
````

````python
# plotting the mean score calculated over the next 100 episodes for each current episode
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores_mean)), scores_mean)
plt.title('DQN agent')
plt.ylabel('Score for the last 100 episodes')
plt.xlabel('Episode #')
plt.show()
````

````python
# plotting MSE loss calculated for each time step  
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(agent.losses)), agent.losses)
plt.title('DQN agent')
plt.ylabel('Losses / timestep')
plt.xlabel('timestep')
plt.show()
````

## DoubleDQN

````python
from DoubleDQN import DoubleDQNAgent
````

````python
ddqagent = DoubleDQNAgent(state_size=37, action_size=4, seed=0)
````

````python
def doubledqn(n_episodes = 2000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995):
        start = timer()    # timer starts for training
        scores = []        # list to contain the scores from each episode
        scores_window = deque(maxlen=100)   # last 100 scores
        scores_mean = []                    # list that scores mean of last 100 episodes
        eps = eps_start
        
        for i_episode in range(1, n_episodes+1):
            state = env.reset(train_mode=True)[brain_name].vector_observations[0]
            score = 0
            for t in range(max_t):
                
                action = ddqagent.act(state, eps)
                env_info= env.step(action)[brain_name]
                next_state = env_info.vector_observations[0]
                reward = env_info.rewards[0]
                done = env_info.local_done[0]
                
                ddqagent.step(state, action, reward, next_state, done)
                
                state = next_state
                score += reward
                if done:
                    break
            scores_window.append(score)    # save most recent score
            scores.append(score)           # save most recent score
            eps = max(eps_end, eps_decay*eps)  # decrease epsilon
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
            
            if i_episode % 100 == 0:
                print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            if i_episode >= 100:
                scores_mean.append(np.mean(scores_window))
            if np.mean(scores_window) >= 13.0:     # environment solved
                print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
                torch.save(ddqagent.qnetwork_online.state_dict(), 'checkpoint_doubledqn.pth')
                break

        return scores, scores_mean, ddqagent
````

````python
dscores, dscores_mean, ddqagent = doubledqn()        
````

````python
# plotting per episode score achieved by the agent

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(dscores)), dscores)
plt.title('Double DQN agent')
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
````

````python
# plotting the mean score calculated over the next 100 episodes for each current episode

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(dscores_mean)), dscores_mean)
plt.title('Double DQN agent')
plt.ylabel('Score for the last 100 episodes')
plt.xlabel('Episode #')
plt.show()
````
